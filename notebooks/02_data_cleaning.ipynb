# Data Cleaning

## Import relevant libraries

```python
import pandas as pd
import numpy as np
```

## Load Interim Data

```python
train_df = pd.read_csv("../data/interim/train_window.csv", parse_dates=["InvoiceDate"])
validation_df = pd.read_csv("../data/interim/validation_window.csv")

print("Train shape:", train_df.shape)
print("Validation shape:", validation_df.shape)
```

## Change InvoiceData data type

```python
train_df["InvoiceDate"] = pd.to_datetime(
    train_df["InvoiceDate"],
    format="mixed",
    dayfirst=True
)

validation_df["InvoiceDate"] = pd.to_datetime(
    validation_df["InvoiceDate"],
    format="mixed",
    dayfirst=True
)
```

```python
train_df.info()
```

```python
validation_df.info()
```

### Strip Column Names

```python
train_df.columns = train_df.columns.str.strip()
validation_df.columns = validation_df.columns.str.strip()
```

## Remove Duplicate Rows

```python
train_df = train_df.drop_duplicates()
validation_df = validation_df.drop_duplicates()
```

## Remove Cancelled Invoices

```python
train_df = train_df[~train_df["Invoice"].astype(str).str.startswith("C")]
validation_df = validation_df[~validation_df["Invoice"].astype(str).str.startswith("C")]
```

## Remove Invalid Quantity & Price

```python
train_df = train_df[(train_df["Quantity"] > 0) & (train_df["Price"] > 0)]
validation_df = validation_df[(validation_df["Quantity"] > 0) & (validation_df["Price"] > 0)]
```

## Remove Missing Customer ID

```python
train_df = train_df.dropna(subset=["Customer ID"])
validation_df = validation_df.dropna(subset=["Customer ID"])
```

## Convert Customer ID to String

```python
train_df["Customer ID"] = train_df["Customer ID"].astype(int).astype(str)
validation_df["Customer ID"] = validation_df["Customer ID"].astype(int).astype(str)
```

## Create TotalPrice Feature

```python
train_df["TotalPrice"] = train_df["Quantity"] * train_df["Price"]
validation_df["TotalPrice"] = validation_df["Quantity"] * validation_df["Price"]
```

```python
train_df.info()
train_df.describe()
```

## Reset Index

```python
train_df = train_df.reset_index(drop=True)
validation_df = validation_df.reset_index(drop=True)
```

## Save Cleaned Datasets

```python
train_df.to_csv("../data/processed/train_clean.csv", index=False)
validation_df.to_csv("../data/processed/validation_clean.csv", index=False)
```

## Final Summary – Data Cleaning Completed

In this notebook, I prepared the interim datasets for feature engineering and segmentation analysis.

The goal of this stage was to ensure that only valid and meaningful transactional data is used for RFM calculation.

### Cleaning steps performed:

- Removed duplicate records

- Removed cancelled invoices (Invoice numbers starting with "C")

- Removed transactions with non-positive Quantity or Price

- Removed records with missing Customer ID

- Customer ID from float to string to correctly represent it as an identifier

- Parsed InvoiceDate as datetime format

- Created a new feature: TotalPrice = Quantity × Price

After applying these steps, the datasets now contain only valid purchase transactions linked to identifiable customers.
